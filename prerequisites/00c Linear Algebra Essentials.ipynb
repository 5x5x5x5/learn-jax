# 00c Linear Algebra Essentials

Neural networks are built from linear algebra operations — matrix multiplications, dot products, and transformations. You don't need to be a mathematician, but understanding what these operations do (and what shapes they produce) will make JAX code intuitive rather than mysterious.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Vectors

A vector is a 1D array. The dot product measures similarity and is the core of every neuron.

```python
a = np.array([1.0, 2.0, 3.0])
b = np.array([4.0, 5.0, 6.0])

# Basic operations
print("a + b:", a + b)
print("2 * a:", 2 * a)

# Dot product: sum of element-wise products
print("dot(a, b):", np.dot(a, b))  # 1*4 + 2*5 + 3*6 = 32
print("manual:", np.sum(a * b))     # Same thing
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
origin = [0, 0]
a = np.array([2, 1])
b = np.array([1, 2])
ax.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, color='blue', label='a')
ax.quiver(*origin, *b, angles='xy', scale_units='xy', scale=1, color='red', label='b')
ax.quiver(*origin, *(a + b), angles='xy', scale_units='xy', scale=1, color='purple', label='a + b')
ax.set_xlim(-0.5, 4.5)
ax.set_ylim(-0.5, 4.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title("Vector Addition")
plt.tight_layout()
plt.show()
```

## Matrices and matrix multiplication

A matrix transforms vectors. A neural network layer is just a matrix multiply.

```python
# Matrix-vector multiplication: transform a vector
W = np.array([[1, 2], [3, 4], [5, 6]])  # (3, 2)
x = np.array([1.0, 2.0])                 # (2,)
print("W @ x:", W @ x)                    # (3,) — transforms 2D input to 3D output
print("Shape: (3,2) @ (2,) → (3,)")

# Matrix-matrix multiplication
A = np.random.randn(3, 4)
B = np.random.randn(4, 2)
C = A @ B
print(f"\n({A.shape}) @ ({B.shape}) → ({C.shape})")
print("Rule: inner dimensions must match, result is (outer dims)")
```

## The neural network connection

```python
# A single neuron: y = dot(weights, input) + bias
weights = np.array([0.5, -0.3, 0.8])
input_features = np.array([1.0, 2.0, 3.0])
bias = 0.1

output = np.dot(weights, input_features) + bias
print(f"Single neuron output: {output:.2f}")

# A full layer: Y = X @ W + b (all inputs transformed at once)
X = np.random.randn(4, 3)   # 4 examples, 3 features each
W = np.random.randn(3, 2)   # Transform 3 features → 2 outputs
b = np.array([0.1, -0.1])   # One bias per output

Y = X @ W + b                # (4,3) @ (3,2) + (2,) → (4,2)
print(f"\nInput shape:  {X.shape}  (4 examples, 3 features)")
print(f"Weight shape: {W.shape}  (3 inputs, 2 outputs)")
print(f"Output shape: {Y.shape}  (4 examples, 2 outputs)")
```

## Transpose

Flips rows and columns. Needed for aligning shapes.

```python
A = np.arange(6).reshape(2, 3)
print("A:\n", A)
print(f"A shape: {A.shape}")

print("\nA.T:\n", A.T)
print(f"A.T shape: {A.T.shape}")

# Common pattern: X.T @ X gives a (features × features) matrix
X = np.random.randn(100, 3)
print(f"\nX.T @ X shape: {(X.T @ X).shape}")  # (3, 3)
```

## Norms

Measuring vector length. Used in regularization and normalization.

```python
x = np.array([3.0, 4.0])

# L2 norm (Euclidean length)
print("L2 norm:", np.linalg.norm(x))  # sqrt(9 + 16) = 5

# Normalizing a vector (unit length)
x_normalized = x / np.linalg.norm(x)
print("Normalized:", x_normalized)
print("Length after normalizing:", np.linalg.norm(x_normalized))  # 1.0
```

## Element-wise vs matrix operations

This distinction trips up many beginners.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Element-wise multiply (A * B):\n", A * B)
print("\nMatrix multiply (A @ B):\n", A @ B)
print("\nThey are NOT the same!")
```

## Summary

| Operation | Code | Shape rule | ML use case |
|---|---|---|---|
| Dot product | `np.dot(a, b)` | (n,) · (n,) → scalar | Single neuron |
| Matrix-vector | `W @ x` | (m,n) @ (n,) → (m,) | One layer, one input |
| Matrix-matrix | `A @ B` | (m,n) @ (n,p) → (m,p) | One layer, batch of inputs |
| Transpose | `A.T` | (m,n) → (n,m) | Aligning shapes |
| Norm | `np.linalg.norm(x)` | → scalar | Regularization, normalization |
| Element-wise | `A * B` | Same shapes | Gradient scaling, masking |
