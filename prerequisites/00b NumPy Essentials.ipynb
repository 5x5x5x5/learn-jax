# 00b NumPy Essentials

JAX's array API (`jax.numpy`) mirrors NumPy almost exactly. Understanding NumPy's core operations — array creation, indexing, broadcasting, and reductions — will make JAX feel immediately familiar.

```python
import numpy as np
```

## Array creation

```python
a = np.array([1, 2, 3])
print(f"array: {a}, shape: {a.shape}, dtype: {a.dtype}")

print(np.zeros((2, 3)))
print(np.ones((2, 3)))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))

# Random arrays
rng = np.random.default_rng(42)
print(rng.normal(size=(2, 3)))
```

## Indexing and slicing

```python
a = np.arange(12).reshape(3, 4)
print("Array:\n", a)

print("Row 0:", a[0])
print("Col 1:", a[:, 1])
print("Slice:", a[0:2, 1:3])
print("Last row:", a[-1])

# Boolean indexing
print("Positive:", a[a > 5])
```

## Shape manipulation

```python
a = np.arange(12)
print(a.reshape(3, 4))
print(a.reshape(3, 4).T)  # Transpose: (3,4) -> (4,3)

# Stacking
x = np.array([1, 2, 3])
print(np.stack([x, x, x]))  # (3, 3)

# Adding dimensions for broadcasting
print(x.shape)                    # (3,)
print(x[np.newaxis, :].shape)     # (1, 3) — row vector
print(x[:, np.newaxis].shape)     # (3, 1) — column vector
```

## Broadcasting

This is critical for JAX. Matrix operations and adding biases to batches rely on broadcasting.

```python
# Scalar + array
print(np.array([1, 2, 3]) + 10)

# Column + row → matrix (outer product pattern)
col = np.array([[1], [2], [3]])    # (3, 1)
row = np.array([[10, 20, 30]])     # (1, 3)
print(col + row)                    # (3, 3)

# Adding bias to a batch — the most common ML pattern
batch = np.ones((4, 3))  # 4 examples, 3 features
bias = np.array([10, 20, 30])  # (3,)
print((batch + bias).shape)  # (4, 3) — bias added to each row
```

## Reductions

Used in loss functions and normalization.

```python
a = np.arange(12).reshape(3, 4)
print("Array:\n", a)

print("Sum all:", np.sum(a))
print("Sum rows (axis=0):", np.sum(a, axis=0))   # shape (4,)
print("Sum cols (axis=1):", np.sum(a, axis=1))    # shape (3,)
print("Mean per col:", np.mean(a, axis=0))

# keepdims preserves the reduced axis as size 1
print("Sum with keepdims:", np.sum(a, axis=1, keepdims=True).shape)  # (3, 1)
```

## Element-wise ops and linear algebra

```python
a = np.array([1.0, 2.0, 3.0])
b = np.array([4.0, 5.0, 6.0])

# Element-wise
print("a * b:", a * b)
print("a + b:", a + b)
print("exp(a):", np.exp(a))

# Dot product (scalar result)
print("dot:", np.dot(a, b))

# Matrix multiplication
A = np.random.randn(3, 4)
B = np.random.randn(4, 2)
print("A @ B shape:", (A @ B).shape)  # (3, 2)
```

## Where and masking

```python
a = np.array([-2, -1, 0, 1, 2])

# Replace negatives with 0 (like ReLU!)
print(np.where(a > 0, a, 0))

# Conditional selection
print(np.where(a % 2 == 0, "even", "odd"))
```

## Summary

| Operation | NumPy | JAX equivalent |
|---|---|---|
| Create array | `np.array(...)` | `jnp.array(...)` |
| Zeros/ones | `np.zeros(shape)` | `jnp.zeros(shape)` |
| Reshape | `a.reshape(m, n)` | `a.reshape(m, n)` |
| Sum/mean | `np.sum(a, axis=0)` | `jnp.sum(a, axis=0)` |
| Matrix multiply | `A @ B` | `A @ B` |
| Conditional | `np.where(cond, x, y)` | `jnp.where(cond, x, y)` |
| Broadcasting | Automatic | Automatic |
