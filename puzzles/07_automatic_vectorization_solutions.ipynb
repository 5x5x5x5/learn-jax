## Prerequisites

JAX installed via `uv sync`

```python
import jax
import jax.numpy as jnp
```

```python
data = jnp.arange(12).reshape(3, 4)

def scale(x):
    return x * 2.0

def dot_product(a, b):
    return jnp.sum(a * b)
```

### Exercise 61

Use `jax.vmap` to apply `scale` to each row of `data` and assign to `scaled_rows`.

```python
scaled_rows = jax.vmap(scale)(data)
scaled_rows
```

### Exercise 62

Use `jax.vmap` with `in_axes=1` to apply `scale` to each column of `data` and assign to `scaled_cols`.

```python
scaled_cols = jax.vmap(scale, in_axes=1)(data)
scaled_cols
```

### Exercise 63

Compute dot product of each row of `data` with `jnp.array([1., 2., 3., 4.])` using vmap. Assign to `dots`.

```python
v = jnp.array([1., 2., 3., 4.])
dots = jax.vmap(lambda row: dot_product(row, v))(data.astype(jnp.float32))
dots
```

### Exercise 64

Use `in_axes=(0, None)` to broadcast the vector across rows. Assign to `dots_broadcast`.

```python
v = jnp.array([1., 2., 3., 4.])
dots_broadcast = jax.vmap(dot_product, in_axes=(0, None))(data.astype(jnp.float32), v)
dots_broadcast
```

### Exercise 65

Create batched weight matrices (5, 4, 3), inputs (5, 4), shared bias (3,). Use vmap with `in_axes=(0, 0, None)` to compute `W @ x + b`. Assign to `batch_outputs`.

```python
key = jax.random.key(0)
W_batch = jax.random.normal(key, (5, 4, 3))
x_batch = jax.random.normal(key, (5, 4))
b = jnp.zeros(3)
def linear(W, x, b): return W @ x + b
batch_outputs = jax.vmap(linear, in_axes=(0, 0, None))(W_batch, x_batch, b)
batch_outputs.shape
```

### Exercise 66

Use nested vmap to compute pairwise dot products between A (3,4) and B (5,4). Assign to `pairwise_dots`.

```python
A = jnp.arange(12, dtype=jnp.float32).reshape(3, 4)
B = jnp.arange(20, dtype=jnp.float32).reshape(5, 4)
pairwise_dots = jax.vmap(lambda a: jax.vmap(lambda b: dot_product(a, b))(B))(A)
pairwise_dots
```

### Exercise 67

Compose `jax.jit(jax.vmap(scale))` and assign to `fast_batch_scale`. Run on `data`.

```python
fast_batch_scale = jax.jit(jax.vmap(scale))
fast_batch_scale(data)
```

### Exercise 68

Use `jax.vmap(jax.grad(...))` to compute per-row gradient of sum-of-squares. Assign to `batch_grads`.

```python
batch_grads = jax.vmap(jax.grad(lambda x: jnp.sum(x ** 2)))(data.astype(jnp.float32))
batch_grads
```

### Exercise 69

Stack a list of dicts into a batched pytree and vmap a function computing `d["a"] + d["b"]`. Assign to `pytree_results`.

```python
batch = [{"a": jnp.array([1., 2.]), "b": jnp.array([3., 4.])}, {"a": jnp.array([5., 6.]), "b": jnp.array([7., 8.])}]
stacked = jax.tree.map(lambda *xs: jnp.stack(xs), *batch)
pytree_results = jax.vmap(lambda d: d["a"] + d["b"])(stacked)
pytree_results
```

### Exercise 70

Use `out_axes=1` so batch dim is axis 1. Assign to `scaled_colwise`.

```python
scaled_colwise = jax.vmap(scale, out_axes=1)(data)
scaled_colwise
```
