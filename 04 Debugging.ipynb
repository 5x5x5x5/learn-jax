# 04 Debugging

Debugging JIT-compiled and vectorized code is different from debugging regular Python. Standard `print()` and `pdb` don't work with traced values. JAX provides specialized tools — `jax.debug.print()`, `jax.debug.breakpoint()`, and `jax.debug.callback()` — that work correctly inside all JAX transformations.

Original Documentation: https://docs.jax.dev/en/latest/debugging.html

```python
import jax
import jax.numpy as jnp
```

## Print

`jax.debug.print()` can be used for traced array values within `jax.jit`, `jax.vmap`, and others. Python `print()` will not work here because it will only execute at trace-time (and will print out the abstract tracer object types instead).

Python `print()` should be used otherwise for static values such as dtypes and array shapes.


```python
@jax.jit
def f(x):
    jax.debug.print("x={x}", x=x)
    y = jnp.sin(x)
    jax.debug.print("y={y}", y=y)
    return y


f(2.0)
```

Similarly, with `jax.vmap` only `jax.debug.print()` will print the values being mapped over.


```python
@jax.vmap
def f(x):
    jax.debug.print("x={x}", x=x)
    y = jnp.sin(x)
    jax.debug.print("y={y}", y=y)
    return y


xs = jnp.arange(3)
f(xs)
```

Note that all `x` prints occur first, followed by all `y` prints.

Instead, if we want sequential prints, we can use `jax.lax.map()`:


```python
from jax import lax


def f(x):
    jax.debug.print("x={x}", x=x)
    y = jnp.sin(x)
    jax.debug.print("y={y}", y=y)
    return y


xs = jnp.arange(3)
lax.map(f, xs)
```

Notice now that the ordering of the prints is not sequential (i.e., `x` is not printed before `y`). This is a byproduct of JAX deeming that `y` depends on `x`.

If we need the original order, we can add `ordered=True`:


```python
def f(x):
    jax.debug.print("x={x}", x=x, ordered=True)
    y = jnp.sin(x)
    jax.debug.print("y={y}", y=y, ordered=True)
    return y


xs = jnp.arange(3)
lax.map(f, xs)
```

## Breakpoint

Use `jax.debug.breakpoint()` to get PDB-style debugging that will pause program execution.


```python
@jax.jit
def f(x):
    jax.debug.breakpoint()
    y, z = jnp.sin(x), jnp.cos(x)
    jax.debug.breakpoint()
    return y, z


f(2.0)
```

For value-dependent breakpoints, we can use `jax.lax.select()`:


```python
def break_if_nonzero(x):
    def true_fun(x):
        pass

    def false_fun(x):
        jax.debug.breakpoint()

    jax.lax.cond(x == 0, true_fun, false_fun, x)


@jax.jit
def f(x):
    break_if_nonzero(x)
    return jnp.sin(x)


f(2.0)  # Non-zero input
```

## Callbacks

If we need more control, we can use `jax.debug.callback()` which gives greater control over the host-side logic via a Python callback. It is compatible with every transformation (`jax.vmap()`, `jax.jit()`, etc.):


```python
import logging


def log_x(x):
    logging.warning(f"Logging value: {x}")


@jax.jit
def f(x):
    jax.debug.callback(log_x, x)
    return jnp.cos(x)


f(2.0)
```

Here’s an example in `jax.vmap()` transformation:


```python
def log_x(x):
    logging.warning(f"Logging value: {x}")


@jax.jit
def f(x):
    jax.debug.callback(log_x, x)
    return jnp.cos(x)


batch_f = jax.vmap(f, in_axes=0)

xs = jnp.stack(jnp.arange(3))
batch_f(xs)
```

## When to use what

| Tool | Best for | Works in JIT | Works in vmap | Interactive |
|---|---|---|---|---|
| `jax.debug.print()` | Quick inspection of traced values | Yes | Yes | No |
| `jax.debug.breakpoint()` | Step-through debugging, inspecting state | Yes | Limited | Yes |
| `jax.debug.callback()` | Custom host-side logic (logging, recording) | Yes | Yes | No |
| Python `print()` | Static values (shapes, dtypes) at trace time | Trace-time only | Trace-time only | No |

**Tips:**
- Start with `jax.debug.print()` — it covers most cases
- Use `ordered=True` when print ordering matters (costs some performance)
- Use `jax.debug.callback()` when you need to call arbitrary Python (e.g., `logging`, writing to a file)
- Use `jax.debug.breakpoint()` sparingly — it pauses execution and is best for targeted investigation

## Summary

| Function | Purpose |
|---|---|
| `jax.debug.print("x={x}", x=val)` | Print traced values inside JIT/vmap |
| `jax.debug.print(..., ordered=True)` | Print with guaranteed sequential ordering |
| `jax.debug.breakpoint()` | Pause execution for interactive PDB-style debugging |
| `jax.debug.callback(fn, *args)` | Call a Python function with traced values on the host |
| `jax.lax.map(f, xs)` | Sequential (non-vectorized) map, useful for ordered debugging |
